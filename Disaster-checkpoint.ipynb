```python
#Milestone 1: Data Acquisition, Cleaning & Initial Exploration
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Insert the csv file
df = pd.read_csv("database.csv.csv")
df.head()
```

```python
# showing the dataframe in the dataset and check structure of the dataset
df.info()
```

```python
#describing the information of dataset for checking basic statistics
df.describe()
```

```python
#labeling the columns which contain spaces or uppercase letters.
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
```

```python
#checking the null or missing values in the dataset
df.isnull().sum()
```

```python
#remove the row with missing values from the dataset 
df.dropna(how="all", inplace=True)
```

```python
#defining the duplicate values
df.duplicated().sum()
```

```python
#droping the duplicate values from the dataset
df.drop_duplicates(inplace=True)
```

```python
# convert date columns
df["declaration_date"] = pd.to_datetime(df["declaration_date"])
df["start_date"] = pd.to_datetime(df["start_date"])
df["end_date"] = pd.to_datetime(df["end_date"])
df["close_date"] = pd.to_datetime(df["close_date"])
```

```python
#checking the data type again
df.dtypes
```

```python
#disaster type formatting
df["disaster_type"] = df["disaster_type"].astype(str).str.strip().str.title()
```

```python
#state codes uppercase
df["state"] = df["state"].str.upper()
```

```python
#checking the unique categories
df["disaster_type"].unique()
```

```python
#creating the new time columns for analysis
df["year"] = df["declaration_date"].dt.year
df["month"] = df["declaration_date"].dt.month
df["day"] = df["declaration_date"].dt.day
```

```python
#frequency count - declarations per state
state_counts = df["state"].value_counts()

print(state_counts)
```

```python
#defining the most common disaster types
disaster_counts = df["disaster_type"].value_counts()

print(disaster_counts)
```

```python
# Time Series plot for the Declarations per Year
year_counts = df["year"].value_counts().sort_index()

plt.plot(year_counts.index, year_counts.values)
plt.xlabel("Year")
plt.ylabel("Number of Declarations")
plt.title("Declarations per Year")
plt.show()
```

```python
# Bar Chart for Declarations per State
state_counts = df["state"].value_counts()
state_counts.plot(kind="bar",figsize=(10,5))
plt.title("Declaration per State")
plt.xlabel("state")
plt.ylabel("Count")
state_counts.head(15).plot(kind="bar")
plt.show()
```

```python
year_incident = pd.crosstab(df["year"], df["disaster_type"])

year_incident.plot(figsize=(12,6))

plt.title("Disaster Types by Year")
plt.xlabel("Year")
plt.ylabel("Count")

plt.show()
```

```python
#Milestone 2: Temporal Trend Visualization
```

```python
#Refined plot - total declarations per year
yearly_counts = df.groupby("year").size()

plt.figure(figsize=(10,5))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o')
plt.title("Total Disaster Declarations Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Declarations")
plt.grid(True)
plt.show()
```

```python
#incident type trends over time(Multiple line chart)
incident_trends = df.groupby(["year","disaster_type"]).size().unstack()

incident_trends.plot(figsize=(12,6))
plt.title("Incident Type Trends Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Declarations")
plt.legend(title="Incident Type")
plt.show()
```

```python
#Seasonality Analysis(Average declarations per month)
monthly_counts = df.groupby("month").size()

plt.figure(figsize=(8,5))
sns.barplot(x=monthly_counts.index, y=monthly_counts.values)

plt.title("Average Disaster Declarations Per Month")
plt.xlabel("Month")
plt.ylabel("Number of Declarations")
plt.show()
```

```python
#stracked area chart for disaster types over time
incident_trends = df.groupby(["year","disaster_type"]).size().unstack().fillna(0)

incident_trends.plot.area(figsize=(12,6))

plt.title("Disaster Types Over Time (Stacked Area)")
plt.xlabel("Year")
plt.ylabel("Number of Declarations")
plt.show()
```

```python
print(df.columns)
```

```python
#incident duration analysis(if start and end dates exist)
df["start_date"] = pd.to_datetime(df["start_date"])
df["end_date"] = pd.to_datetime(df["end_date"])

df["duration_days"] = (df["end_date"] - df["start_date"]).dt.days
```

```python
#Average duration per year
duration_year = df.groupby("year")["duration_days"].mean()

plt.figure(figsize=(10,5))
plt.plot(duration_year.index, duration_year.values, marker='o')

plt.title("Average Disaster Duration Over Time")
plt.xlabel("Year")
plt.ylabel("Average Duration (Days)")
plt.show()
```

```python
#showing spatial distribution of disasters across states
import plotly.express as px

state_counts = df.groupby("state").size().reset_index(name="declarations")

fig = px.choropleth(
    state_counts,
    locations="state",
    locationmode="USA-states",
    color="declarations",
    scope="usa",
    color_continuous_scale="Reds",
    title="Total Disaster Declarations by State"
)

fig.show()
```

```python
#bar chart comparing states
#helps identify states most affected by disaster.
top_states = df["state"].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_states.index,
    y=top_states.values
)

plt.title("Top 10 States with Most Disaster Declarations")
plt.xlabel("State")
plt.ylabel("Number of Declarations")

plt.show()
```

```python
#disaster type distribution by state(e.g Hurricanes)
#shows specific disaster hotspots
hurricane_data = df[df["disaster_type"] == "Hurricane"]

hurricane_states = hurricane_data["state"].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=hurricane_states.index,
    y=hurricane_states.values
)

plt.title("States Most Affected by Hurricanes")
plt.xlabel("State")
plt.ylabel("Number of Hurricane Declarations")

plt.show()
```

```python
#heatmap(state vs disaster type)
#shows which states experience which disasters
state_incident = pd.crosstab(df["state"], df["disaster_type"])

plt.figure(figsize=(12,8))

sns.heatmap(state_incident, cmap="YlOrRd")

plt.title("Disaster Type Distribution Across States")

plt.show()
```

```python
#choropleth map for specific disaster type(e.g Floods)
#highlights flood-prone area
flood_data = df[df["disaster_type"] == "Flood"]

flood_states = flood_data.groupby("state").size().reset_index(name="flood_count")

fig = px.choropleth(
    flood_states,
    locations="state",
    locationmode="USA-states",
    color="flood_count",
    scope="usa",
    color_continuous_scale="Blues",
    title="Flood Disaster Declarations by State"
)

fig.show()
```

```python
#Stacked bar chart(state vs disaster type)
#shows disaster composition in each state
state_incident = pd.crosstab(df["state"], df["disaster_type"])

state_incident.head(10).plot(
    kind="bar",
    stacked=True,
    figsize=(12,6)
)

plt.title("Disaster Types per State")
plt.xlabel("State")
plt.ylabel("Number of Declarations")

plt.show()
```

```python
#overall frequency distribution of incident types
#this shows which disasters occur most often
incident_counts = df["disaster_type"].value_counts()

plt.figure(figsize=(10,5))

sns.barplot(
    x=incident_counts.index,
    y=incident_counts.values
)

plt.xticks(rotation=45)

plt.title("Frequency of Disaster Incident Types")
plt.xlabel("Disaster Type")
plt.ylabel("Number of Declarations")

plt.show()
```

```python
#Treemap shows disaster proportions.
#Larger rectangles represent more frequent disasters.
import squarify

sizes = df["disaster_type"].value_counts()

plt.figure(figsize=(10,6))

squarify.plot(
    sizes=sizes.values,
    label=sizes.index,
    alpha=0.8
)

plt.title("Treemap of Disaster Incident Types")

plt.axis("off")

plt.show()
```

```python
#Relationship Between Incident Type and State
#The stacked bar chart shows which disasters affect each state.
state_incident = pd.crosstab(df["state"], df["disaster_type"])

state_incident.head(10).plot(
    kind="bar",
    stacked=True,
    figsize=(12,6)
)

plt.title("Incident Types by State")
plt.xlabel("State")
plt.ylabel("Number of Declarations")

plt.legend(title="Incident Type")

plt.show()
```

```python
#Assistance Programs vs Incident Type
#Dataset usually includes:
#ihProgramDeclared
#paProgramDeclared
#These show whether aid programs were declared.
#this shows
#Which disasters require housing assistance
#Example: Hurricanes and floods often require housing aid.
ih_analysis = pd.crosstab(df["disaster_type"], df["individual_assistance_program"])

ih_analysis.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Individual Housing Assistance by Incident Type")
plt.xlabel("Incident Type")
plt.ylabel("Count")

plt.show()
```

```python
#Shows how government infrastructure support relates to disasters.
pa_analysis = pd.crosstab(df["disaster_type"], df["public_assistance_program"])

pa_analysis.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Public Assistance Programs by Incident Type")
plt.xlabel("Incident Type")
plt.ylabel("Count")

plt.show()
```

```python
#Heatmap for Incident Type vs State
#Darker color = more disasters
#Helps quickly identify regional disaster patterns.
state_incident = pd.crosstab(df["state"], df["disaster_type"])

plt.figure(figsize=(12,8))

sns.heatmap(
    state_incident,
    cmap="YlOrRd"
)

plt.title("Heatmap of Incident Types by State")

plt.show()
```

```python
# Pie Chart for Disaster Distribution
# Shows percentage share of each disaster type
incident_counts = df["disaster_type"].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    incident_counts.values,
    labels=incident_counts.index,
    autopct="%1.1f%%"
)

plt.title("Disaster Type Distribution")

plt.show()
```
