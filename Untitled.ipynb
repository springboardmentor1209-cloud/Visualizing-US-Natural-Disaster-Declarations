```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#style
sns.set(style="whitegrid")
```

```python
#Load data 
df=pd.read_csv("us_disaster_declarations.csv")
```

```python
df.head()
```

```python
# Missing Values
df.isnull().sum()
```

```python
#date column convert
df['declaration_date']=pd.to_datetime(df['declaration_date'],errors='coerce')
```

```python
#year extract
df['year']=df['declaration_date'].dt.year
```

```python
#duplicates remove
df=df.drop_duplicates()
```

```python
#cleaned data check
df.head()
```

```python
#dataset info
df.info()
```

```python
#summary statistics 
df.describe()
```

```python
#disaster type count
df['incident_type'].value_counts()
```

```python
#top states with most disasters
df['state'].value_counts().head(10)
```

```python
#Visualization
#Year wise Trend
yearly=df['year'].value_counts().sort_index()

plt.figure()
yearly.plot()
plt.title("Year-wise Disaster Trend")
plt.xlabel("Year")
plt.ylabel("Number of Disasters")
plt.show()
```

```python
#Top 10 States

top_states=df['state'].value_counts().head(10)
plt.figure()
top_states.plot(kind='bar')
plt.title("Top 10 States with Most Disasters")
plt.xlabel("State")
plt.ylabel("count")
plt.show()
```

```python
#States vs Disaster Type

cross_tab=pd.crosstab(df['state'],df['incident_type'])

cross_tab.head()
```
